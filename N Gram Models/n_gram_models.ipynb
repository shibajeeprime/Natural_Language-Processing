## Tokenizing

## Vocabulary

```python
vocab = set(token_sentence)
```

```python
len(vocab)
```

## n gram counts

```python
unigram_counts = {}
bigram_counts = {}
trigram_counts = {}
fourgram_counts= {}
fivegram_counts = {}

for i in range(len(token_sentence)-4):
    unigram = (token_sentence[i])
    bigram = (token_sentence[i], token_sentence[i+1])
    trigram = (token_sentence[i], token_sentence[i+1], token_sentence[i+2])
    fourgram = (token_sentence[i], token_sentence[i+1], token_sentence[i+2], token_sentence[i+3])
    fivegram = (token_sentence[i], token_sentence[i+1], token_sentence[i+2], token_sentence[i+3], token_sentence[i+4])
    
    
    if unigram in unigram_counts.keys():
        unigram_counts[unigram] += 1
    else:
        unigram_counts[unigram] = 1
    
    
    
    if bigram in bigram_counts.keys():
        bigram_counts[bigram] += 1
    else:
        bigram_counts[bigram] = 1
        
        
        
    if trigram in trigram_counts.keys():
        trigram_counts[trigram] += 1
    else:
        trigram_counts[trigram] = 1
        
        
    
    if fourgram in fourgram_counts.keys():
        fourgram_counts[fourgram] += 1
    else:
        fourgram_counts[fourgram] = 1
        
        
        
    if fivegram in fivegram_counts.keys():
        fivegram_counts[fivegram] += 1
    else:
        fivegram_counts[fivegram] = 1
```

## Unigram Model

```python
def next_word_uni(inputs):
    tokenize_input = inputs.lower().split()
    last = tokenize_input[-1:]
    
    vocab_prob = {}
    for vocab_word in vocab:
        test_bigram = (last[0],  vocab_word)
        test_unigram = (last[0])
        
        test_bigram_count = bigram_counts.get(test_bigram, 0)
        test_unigram_count = unigram_counts.get(test_unigram, 0)
        
        prob = test_bigram_count/test_unigram_count
        vocab_prob[vocab_word] = prob
    top_suggestions = sorted(vocab_prob.items(), key = lambda x:x[1], reverse = True)[:1]
    return(top_suggestions[0][0])
```

```python
next_word_uni('religious')
```

```python
def unimodel(inputs):
    i = inputs
    k = 0
    while(next_word_uni(i)[-1] != '.'):
        i = i + ' ' + next_word_uni(i)
        k = k+1
        if(k == 50):
            break       
    i = i + ' ' + next_word_uni(i)
    return(i)
```

```python
unimodel('spirit')
```

## Bigram Model

```python
def next_word_bi(inputs):
    tokenize_input = inputs.lower().split()
    last = tokenize_input[-2:]
    
    vocab_prob = {}
    for vocab_word in vocab:
        test_trigram = (last[0], last[1], vocab_word)
        test_bigram = (last[0], last[1])
        
        test_trigram_count = trigram_counts.get(test_trigram, 0)
        test_bigram_count = bigram_counts.get(test_bigram, 0)
        
        prob = test_trigram_count/test_bigram_count
        vocab_prob[vocab_word] = prob
    top_suggestions = sorted(vocab_prob.items(), key = lambda x:x[1], reverse = True)[:1]
    return(top_suggestions[0][0])
```

```python
next_word_bi('where the intellect')
```

```python
def bimodel(inputs):
    i = inputs
    while(next_word_bi(i)[-1] != '.'):
        i = i + ' ' + next_word_bi(i)
    i = i + ' ' + next_word_bi(i)
    return(i)
```

```python
bimodel('the possibilities of the soul')
```

## Trigram Model

```python
def next_word_tri(inputs):
    tokenize_input = inputs.lower().split()
    last = tokenize_input[-3:]
    
    vocab_prob = {}
    for vocab_word in vocab:
        test_fourgram = (last[0], last[1], last[2], vocab_word)
        test_trigram = (last[0], last[1], last[2])
        
        test_fourgram_count = fourgram_counts.get(test_fourgram, 0)
        test_trigram_count = trigram_counts.get(test_trigram, 0)
        
        prob = test_fourgram_count/test_trigram_count
        vocab_prob[vocab_word] = prob
    top_suggestions = sorted(vocab_prob.items(), key = lambda x:x[1], reverse = True)[:1]
    return(top_suggestions[0][0])
```

```python
next_word_tri('every day where the')
```

```python
def trimodel(inputs):
    i = inputs
    while(next_word_tri(i)[-1] != '.'):
        i = i + ' ' + next_word_tri(i)
    i = i + ' ' + next_word_tri(i)
    return(i)
```

```python
trimodel('the possibilities of the soul')
```

## Fourgram Model

```python
def next_word_four(inputs):
    tokenize_input = inputs.lower().split()
    last = tokenize_input[-4:]
    
    vocab_prob = {}
    for vocab_word in vocab:
        test_fivegram = (last[0], last[1], last[2], last[3], vocab_word)
        test_fourgram = (last[0], last[1], last[2], last[3])
        
        test_fivegram_count = fivegram_counts.get(test_fivegram, 0)
        test_fourgram_count = fourgram_counts.get(test_fourgram, 0)
        
        prob = test_fivegram_count/test_fourgram_count
        vocab_prob[vocab_word] = prob
    top_suggestions = sorted(vocab_prob.items(), key = lambda x:x[1], reverse = True)[:1]
    return(top_suggestions[0][0])
```

```python
next_word_four('we find cases almost every day where the')
```

```python
def quadmodel(inputs):
    i = inputs
    while(next_word_four(i)[-1] != '.'):
        i = i + ' ' + next_word_four(i)
    i = i + ' ' + next_word_four(i)
    return(i)
```
